```python
import pandas as pd
data = pd.read_csv("../data/superstore.csv")

print("Segment:", data['Segment'].unique())
print("region:", data['Region'].unique())
print("Category", data['Category'].unique())
print("Sub-Category", data['Sub-Category'].unique())
```

```python
sales_by_segment = data.groupby("Segment")["Sales"].sum()
print(sales_by_segment)
```

```python
sales_by_region = data.groupby("Region")["Sales"].sum()
print(sales_by_region)
```

```python
import matplotlib.pyplot as plt

sales_by_segment.plot(kind="bar", color=["skyblue", "orange", "green"])
plt.title("Umsatz nach Kundensegment")
plt.ylabel("Umsatz in $")
plt.xlabel("Segment")
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("../data/superstore.csv")

sales_by_region = data.groupby("Region")["Sales"].sum()
# print(sales_by_region)
sales_by_region.plot(kind="bar", color="coral")
plt.title("Umsatz nach Region")
plt.ylabel("Umsatz in $")
plt.xlabel("Region")
plt.show()
```

```python
sales_by_category = data.groupby("Category")["Sales"].sum()
print(sales_by_category)
sales_by_category.plot(kind="bar", color=["purple", "teal", "gold"])
plt.title("Umsatz nach Produktkategorie")
plt.ylabel("Umsatz in $")
plt.xlabel("Kategorie")
plt.show()
```

```python
sales_by_subcat = data.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
print(sales_by_subcat)

#########visualisierung
sales_by_subcat.plot(kind="bar", figsize=(12,6), color="steelblue")
plt.title("Umsatz nach Sub-Category")
plt.ylabel("Umsatz in $")
plt.xlabel("Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data["Order Date"] = pd.to_datetime(data["Order Date"], dayfirst=True)
sales_over_time = data.groupby(pd.Grouper(key="Order Date", freq="ME"))["Sales"].sum()


sales_over_time.plot(figsize=(12, 6), color="darkorange")
plt.title("Umsatz ber die Zeit")
plt.ylabel("Umsatz in $")
plt.xlabel("Monat")
plt.show()


```

```python
print(" Insights aus der Verkaufsdatenanalyse:")
print("- Das Segment 'Consumer' generiert den hchsten Umsatz (~1,15 Mio $).")
print("- Die umsatzstrkste Region ist X (hier kannst du den Plotwert einsetzen).")
print("- Die Kategorie mit dem hchsten Umsatz ist Technology (siehe Plot).")
print("- Unterkategorie mit dem hchsten Umsatz: z.B. Phones oder Chairs (siehe Sub-Category Plot).")
print("- Umsatz ber die Zeit zeigt saisonale Peaks, z.B. zum Jahresende.")
```
